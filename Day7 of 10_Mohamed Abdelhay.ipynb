```python
df_cop =df.copy()

df_cop=df.dropna()
print ('the Number of NAN value after drop na')
df_cop.isnull().sum()
```

```python
fig, ax = plt.subplots(figsize=(14,7))
sb.heatmap(df.corr(), annot= True,linewidths=0.1);
```

```python
a= 0.23
b= 0.25
df_cop['VP_DT'] =(a * (df_cop['DTC']**b))*1000
df.keys()
df.head(10)
```

```python
plt.figure(figsize=(10,5))
sb.regplot(x=df_cop.VP_DT, y=df_cop.NPHI)
plt.scatter(df_cop.VP_DT, df_cop.NPHI, c = df_cop.GR, cmap = 'rainbow_r', alpha = 0.5)
plt.colorbar()
plt.title('VP_DT vs NPHI plot ')
plt.xlabel('VP_DT')
plt.ylabel('NPHI')
plt.grid();
```
